## 2. Now, a transcription

Someone read the page and wrote down the letters. That's already an interpretation, not a copy: the transcriber decides what each mark is.

```python
beowulf = """Hwæt! We Gardena in geardagum,
þeodcyninga, þrym gefrunon,
hu ða æþelingas ellen fremedon.
Oft Scyld Scefing sceaþena þreatum,
monegum mægþum, meodosetla ofteah,
egsode eorl. Syððan ærest wearð
feasceaft funden, he þæs frofre gebad,
weox under wolcnum, weorðmyndum þah,
oðþæt him æghwylc þara ymbsittendra
ofer hronrade hyran scolde,
gomban gyldan. Þæt wæs god cyning!"""

print(beowulf)
```

## 3. And finally, as data

Reduce the text to characters in a string and the computer can begin to count. Notice what it gains and what it loses.

```python
words = beowulf.replace('\n', ' ').split()
types = set(w.lower().strip('.,!') for w in words)

print('Tokens (running words):', len(words))
print('Types (distinct forms):', len(types))
print('Type/token ratio:', round(len(types) / len(words), 2))
```

A high type/token ratio means little repetition: most words occur once. Verse tends to run high. The same measure on a contract or a legal code runs low.

## 4. When two copies disagree

Line 6 is damaged. The manuscript appears to read *eorl* (one nobleman); most editors print *eorlas* (warriors, plural). That difference is a **variant**, and the cell below shows how a computer represents the disagreement—exactly what Day 2 automates at scale.

```python
import difflib

witness_a = 'egsode eorl'      # what the manuscript seems to show
witness_b = 'egsode eorlas'    # the common emendation

for d in difflib.ndiff(witness_a.split(), witness_b.split()):
    print(d)    # '-' only in A, '+' only in B, ' ' in both
```

## 5. Spelling never settled down

There was no standard spelling. The two scribes of *Beowulf* spell the same words differently, and þ and ð both stand for *th*. To compare forms across copies you usually have to **normalize** first, and every normalization discards some information.

```python
def normalize(s):
    return (s.lower()
             .replace('þ', 'th').replace('ð', 'th')
             .replace('æ', 'ae'))

for w in ['þeodcyninga', 'æþelingas', 'weorðmyndum']:
    print(f'{w:14} ->  {normalize(w)}')
```

## 6. Words have relatives

Set *Beowulf*'s words beside their cousins in related languages and a regular pattern appears. This is where philology became linguistics.

```python
# gloss, Old English, Modern English, German, Latin
cognates = [
    ('father', 'fæder', 'father', 'Vater', 'pater'),
    ('foot',   'fot',   'foot',   'Fuss',  'pes'),
    ('fish',   'fisc',  'fish',   'Fisch', 'piscis'),
    ('full',   'full',  'full',   'voll',  'plenus'),
]
print(f"{'gloss':9}{'OE':9}{'ModE':9}{'German':9}{'Latin':9}")
for row in cognates:
    print(''.join(f'{x:9}' for x in row))

print('\nWhere Latin has p, the Germanic words have f. That regularity is Grimm\'s Law.')
```

## 7. (Stretch) Hear the alliteration

Old English meter is built on alliteration: repeated initial sounds in a line's stressed words. Scanning it properly requires knowing how the words were pronounced, but counting initial letters gives a first indication.

```python
from collections import Counter

line1 = 'Hwæt We Gardena in geardagum'
firsts = [w[0].lower() for w in line1.split()]
print(Counter(firsts))
# 'g' shows up twice (Gardena, geardagum) — the alliterating sound of the line.
```
