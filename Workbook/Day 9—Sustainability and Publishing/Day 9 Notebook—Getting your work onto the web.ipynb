# Day 9—Getting your work onto the web
*Measuring Manuscripts*

Durable publishing comes down to one idea: **static files**. A picture, a table, a page—nothing that has to keep running on a server. This notebook turns your analysis into those files and bundles them for your GitHub Pages repository.

## 1. Save a figure

Whatever your result is, save the picture of it as a PNG. Replace the placeholder plot with your own.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(['a', 'b', 'c'], [3, 7, 4])   # === TO BUILD: your real figure ===
ax.set_title('My result')
fig.savefig('figure.png', dpi=150, bbox_inches='tight')
print('Saved figure.png')
```

## 2. Save a data table

CSV so others can reuse the numbers, HTML so they render on the page.

```python
import pandas as pd
df = pd.DataFrame({'item': ['a', 'b', 'c'], 'count': [3, 7, 4]})   # === TO BUILD: your data ===
df.to_csv('data.csv', index=False)
df.to_html('table.html', index=False, border=0)
print('Saved data.csv and table.html')
df
```

## 3. An interactive chart (optional)

A static PNG is the durable default. For a chart readers can hover over, Plotly writes a self-contained HTML file—still a static file, no server.

```python
try:
    import plotly.express as px
    fig2 = px.bar(df, x='item', y='count', title='My result (interactive)')
    fig2.write_html('chart.html', include_plotlyjs='cdn')
    print('Saved chart.html')
except ImportError:
    print('Plotly not installed. Run  !pip install plotly  if you want this. Optional.')
```

## 4. The page itself

A web page is just a text file. Here is a small, complete one with your title, a sentence, the figure, and the table.

```python
html = '''<!doctype html>
<html lang="en"><head><meta charset="utf-8">
<title>TO BUILD: project title</title>
<style>body{font-family:system-ui,sans-serif;max-width:760px;margin:40px auto;padding:0 16px;line-height:1.5}
img{max-width:100%}table{border-collapse:collapse}td,th{padding:4px 10px;border-bottom:1px solid #ddd}</style>
</head><body>
  <h1>TO BUILD: project title</h1>
  <p>TO BUILD: one paragraph saying what this is and what you found.</p>
  <img src="figure.png" alt="result">
  <h2>Data</h2>
  ''' + open('table.html').read() + '''
  <p><a href="data.csv">Download the data (CSV)</a></p>
</body></html>'''
open('index.html', 'w').write(html)
print('Saved index.html')
```

## 5. A README

The file that lets a stranger—including your future self—understand and rebuild the project. It's the single thing that most often keeps a project alive.

```python
readme = '''# TO BUILD: project title

One paragraph: what this is, who made it, and what it shows.

## Data
Where the data came from, and its format.

## How to rebuild
Open the notebook in Colab and run all cells.

## License
See LICENSE.
'''
open('README.md', 'w').write(readme)
print('Saved README.md')
```

## 6. A license

Without a license, no one legally knows what they may do with your work. CC BY 4.0 is a common, permissive choice for scholarship; it asks only for credit.

```python
license_note = ('This work is licensed under CC BY 4.0.\n'
                'https://creativecommons.org/licenses/by/4.0/\n'
                '=== TO BUILD: add your name and year ===')
open('LICENSE', 'w').write(license_note)
print('Saved LICENSE')
```

## 7. Bundle it for download

Zip the whole site so you can download it in one click, then drop the files into your repository.

```python
import zipfile, os
files = [f for f in ['index.html','figure.png','table.html','data.csv','chart.html','README.md','LICENSE'] if os.path.exists(f)]
with zipfile.ZipFile('my_project_site.zip', 'w') as z:
    for f in files:
        z.write(f)
print('Bundled:', files)
try:
    from google.colab import files as colab_files
    colab_files.download('my_project_site.zip')
except Exception:
    print('(In Colab this would download. Use the Files panel on the left to grab the zip.)')
```
